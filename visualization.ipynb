```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./StudentsPerformance.csv")
print(df['parental level of education'].unique())
print(df.head())
```

```python
# relation of scores and family background

#rotates x axis labels to make them visible

plt.xticks(rotation=90)
main = 'parental level of education'
labels=['some high school', 'high school', 'some college', "associate's degree", "bachelor's degree", "master's degree"]

math_means, reading_means, writing_means = [], [], []
for label in labels:
    # get array of valid scores (in this case math)
    math_scores = df.loc[df[main] == label, 'math score']
    # get mean
    math_mean = sum(math_scores) / len(math_scores)
    # add the sum to math_means
    math_means.append(math_mean)
    
    # same as above, but for reading
    # problem! 
    reading_scores = df.loc[df[main] == label, 'reading score']
    reading_mean = sum(reading_scores) / len(reading_scores)
    reading_means.append(reading_mean)
    
    # same as above, but for writing
    writing_scores = df.loc[df[main] == label, 'writing score']
    writing_mean = sum(writing_scores) / len(writing_scores)
    writing_means.append(writing_mean)
    
x_axis = np.arange(len(labels))

plt.bar(x_axis - 0.2, math_means, 0.2, label = 'Math Scores')
plt.bar(x_axis, reading_means, 0.2, label = 'Reading Scores')
plt.bar(x_axis + 0.2, writing_means, 0.2, label = 'Writing Scores')

plt.xticks(x_axis, labels)
plt.xlabel(main)
plt.ylabel("score")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
